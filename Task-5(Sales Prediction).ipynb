```python
#Data Visulaization

import matplotlib.pyplot as plt
import seaborn as sns
sns.pairplot(dataset,kind="reg")
```

```python
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

### Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

### Training the Multiple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

### Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
sale = regressor.predict([[44.5,39.3,45.1]])
print(sale)
```

### Accuracy

```python
print("Score",regressor.score(X_test, y_test))
```

### Analyzing actual and predicted data using scatter plot

```python
plt.scatter(y_test,y_pred,color='blue',marker='*');
plt.xlabel('Actual');
plt.ylabel('Predicted');
```


### Analyzing actual and predicted data using regplot

```python
sns.regplot(x=y_test,y=y_pred,ci=None,color ='green')
```

### Data processing and spliting
#### Independent and dependent variables

```python
x= dataset.drop("Sales",axis =1)
y=dataset.Sales
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#Creating model
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#Print Intercept and slope
print("Intercept : ", model.intercept_)
print("Slope : ", model.coef_)
```

```python
#Prediction
y_Pred = model.predict(x_test)
y_Pred
```

```python
# Calculation of R2 Score
from sklearn.model_selection import cross_val_score
print(cross_val_score(model, x, y, cv=10, scoring="r2").mean())
```

```python
#Prediction with new values
new_x = [[0,500,300,400]]
print(model.predict(new_x))
```
